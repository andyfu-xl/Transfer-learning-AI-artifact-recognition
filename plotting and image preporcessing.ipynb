```python
list_indexes_classes = []
for temp in os.listdir("F:\imagenet_1k")[1:-1]:
    image_file_names = os.listdir(os.path.join("F:\imagenet_1k", temp))
    indexes_classes = []
    indexes_classes.append(0)
    key_counter = 1
    for i in range(len(image_file_names)):
        if key_counter == 1000:
            break
        if image_file_names[i].startswith(key_list[key_counter]):
            indexes_classes.append(i)
            key_counter += 1
    indexes_classes.append(len(image_file_names))
    list_indexes_classes.append(indexes_classes)
```

## Transfer images from the original dataset
This script was not expected to be used twice, thus the readability is poor

Images were resized before feed into the neural network

```python
c = 0
for j in a:
    class_name = stable_classes[j]
    dst_real = "F:\imagenet_1k_subset\real"
    dst_real_class = os.path.join(directory_real, class_name)
    print(c)
    c += 1
    if len(os.listdir(dst_real_class)) >= 100:
        continue
    for i in range(5):
        temp = os.listdir("F:\imagenet_1k")[1:-1][i]
        image_file_names = os.listdir(os.path.join("F:\imagenet_1k", temp))
        temp_index_classes = list_indexes_classes[i]
        temp_samples = []
        for i in range(temp_index_classes[j], temp_index_classes[j + 1]):
            image_name = image_file_names[i]
            image_path = os.path.join(os.path.join("F:\imagenet_1k", temp), image_name)
            im = Image.open(image_path)
            w, h = im.size
            if (w >= 256) and (h > 256) and (h/w > 2/3) and (h/w < 2/3):
                temp_samples.append(i)
        for sample in temp_samples:
            image_name = image_file_names[sample]
            image_path = os.path.join(os.path.join("F:\imagenet_1k", temp), image_name)
            if len(os.listdir(dst_real_class)) < 100:
                os.system("copy \"%s\" \"%s\""%(image_path, dst_real_class))
# for i in a:
#     src_imagenet = "F:\imagenet_1k"
#     class_name = key_list[i]
#     src_stable_class = os.path.join(dir_stable, class_name)
#     images = os.listdir(src_stable_class)
#     dst_generated = "F:\imagenet_1k_subset\generated"
#     dst_generated_class = os.path.join(directory_generated, class_name)
#     for each in images:
#         src_image = os.path.join(src_stable_class, each)
#         os.system("copy \"%s\" \"%s\""%(src_image, dst_generated_class))
```

## Ploting the graph, the graph was used in the report.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
data1 = pd.read_csv("F:\\resultdata\\convolution_new_val_acc.csv")
data2 = pd.read_csv("F:\\resultdata\\resnet18_val_acc.csv")
data3 = pd.read_csv("F:\\resultdata\\resnet_new_val_acc.csv")
data4 = pd.read_csv("F:\\resultdata\\resnext_new_val_acc.csv")
data5 = pd.read_csv("F:\\resultdata\\swinT_val_acc.csv")
data6 = pd.read_csv("F:\\resultdata\\vanilla_new_val_acc.csv")
zero = pd.Series([0])
fig, ax = plt.subplots()
line1, = ax.plot(pd.concat([zero, data1["convolution_new - val/accuracy"][:80]]).ewm(span = 5).mean(), label='cnn', color="#d0d1e6")
line2, = ax.plot(pd.concat([zero, data2["resnet18_50 - val/accuracy"][:80]]).ewm(span = 5).mean(), label='resnet18', color="#636363")
line3, = ax.plot(pd.concat([zero, data3["resnet_new - val/accuracy"][:80]]).ewm(span = 5).mean(), label='resnext_v1', color="#fed976")
line4, = ax.plot(pd.concat([zero, data4["resnext_new - val/accuracy"][:80]]).ewm(span = 5).mean(), label='resnext_v2', color="#252525")
line5, = ax.plot(pd.concat([zero, data5["sinwT-400 - val/accuracy"][:80]]).ewm(span = 5).mean(), label='swinT', color="#d95f0e")
line6, = ax.plot(pd.concat([zero, data6["vanilla_new - val/accuracy"][:80]]).ewm(span = 5).mean(), label='mlp', color="#253494")
ax.set_ylim([.6, 1])
ax.legend(loc=4)
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.title("Validation Accuracy")
plt.savefig("acc.pdf", format="pdf", bbox_inches="tight")
```
